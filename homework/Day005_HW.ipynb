```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？

import requests
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = r.text

import json
data = json.loads(response)

count = 0
for d in data:
    print(d['title'])
    count += 1

print('==================================')
print('總計回傳',count,'筆資料')
print()
print('欄位名稱如下：')
for name in d:
    print(name)
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」

for d in data:
    print(d['title'], d['createdAt'], d['commentCount'], d['likeCount'])
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」
import statistics 

commentCount = []
likeCount = []
for d in data:
    commentCount.append(d['commentCount'])
    likeCount.append(d['likeCount'])

# Printing the mean 
print("平均留言人數：", statistics.mean(commentCount))
print("平均按讚人數：", statistics.mean(likeCount))
```
