### 比較一下範例檔案中的「File I/O」與「xmltodict」讀出來的內容有什麼差異


File I/O 讀取出來的是整份的XML檔案內容

XNLTODICT 讀取可以直接抓取XML中的一個資料

### 根據範例檔案的結果：

1. 請問高雄市有多少地區有溫度資料？
2. 請取出每一個地區所記錄的第一個時間點跟溫度
3. 請取出第一個地區所記錄的每一個時間點跟溫度

```python
# 1. 請問高雄市有多少地區有溫度資料？

import xmltodict

# 存取檔案

with open("./example/64_72hr_CH.xml", "r", encoding="utf-8") as fd:
    doc = dict(xmltodict.parse(fd.read()))

# 存取我們的資訊
print(doc['cwbopendata']['dataset']['locations']['locationsName'])
print('========================================')

# 用迴圈存取我們的資訊
i=0
for area in doc['cwbopendata']['dataset']['locations']['location']: 
    print(area['locationName'])
    i = i + 1
print('========================================')
print('總計', i,"個行政區")
```

```python
# 2. 請取出每一個地區所記錄的第一個時間點跟溫度

import xmltodict

# 存取檔案

with open("./example/64_72hr_CH.xml", "r", encoding="utf-8") as fd:
    doc = dict(xmltodict.parse(fd.read()))

# 存取我們的資訊
print(doc['cwbopendata']['dataset']['locations']['locationsName'])
print('========================================')

# 用迴圈存取我們的資訊

for area in doc['cwbopendata']['dataset']['locations']['location']: 
    
    for weather in area['weatherElement']:
        for time in weather['time']:
            print(area['locationName'], time['dataTime'], '溫度：', time['elementValue']['value'], "度")
            break
        break

print('========================================')
```

```python
# 3. 請取出第一個地區所記錄的每一個時間點跟溫度

import xmltodict

# 存取檔案

with open("./example/64_72hr_CH.xml", "r", encoding="utf-8") as fd:
    doc = dict(xmltodict.parse(fd.read()))

# 存取我們的資訊

# 用迴圈存取我們的資訊

for area in doc['cwbopendata']['dataset']['locations']['location']: 
    print(doc['cwbopendata']['dataset']['locations']['locationsName'], area['locationName'])
    print('========================================')
    
    for weather in area['weatherElement']:
        for time in weather['time']:
            print(area['locationName'], time['dataTime'], '溫度：', time['elementValue']['value'], "度")
                
    break

print('========================================')
```
